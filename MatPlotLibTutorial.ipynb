# Matplotlib tutorial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# simple plot with 4 number

```python
plt.plot([1,2,3,4])
plt.show()
```

# point x and y value; add title and axis

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Test Plot',fontsize=8,color='g')
plt.xlabel('number n')
plt.ylabel('n^2')
plt.show()
```

# change figure size plot red dots; set axis scales x:0-6 y:0-20

```python
plt.figure(figsize=(2,5))
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.annotate('Square it',(3,6))
plt.show()
```

# Bar charts

```python
plt.clf() #clear figure
x=np.arange(4)
y=[8.8,5.2,3.6,5.9]
plt.xticks(x,('Ankit','Joe','Bill','Flaco'))
plt.bar(x,y,color=['lime','r','k','tan'])

plt.show()
```

# Two set of 10 random dots plotted

```python
d={"Red O":np.random.rand(10),
  "Grn X":np.random.rand(10)}

df=pd.DataFrame(d)
df.plot(style=['ro','gx'])
plt.show()
```

# Time Series- six month of random floats

```python
ts=pd.Series(np.random.randn(180),index=pd.date_range('1/1/2018',periods=180))
df=pd.DataFrame(np.random.randn(180,3),index=ts.index,columns=list("ABC"))
df.cumsum().plot()
plt.show()
```

# Random dots in a scatter

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
sizes=(30*np.random.rand(N))**2 #0 to 15 point radii
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5)
plt.show()
```

# Save figure to image file

```python
plt.figure(figsize=(4,3),dpi=100)
plt.plot([245,170,148,161,196,112,258])
plt.axis([0,7,0,300])
plt.title('Fligt Data')
plt.xlabel("Speed")
plt.savefig('flight.png')
plt.show()
```
