# Non-Linear SVM

```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import svm
```

```python
#load dataset
circle_x,circle_y=datasets.make_circles(n_samples=300,noise=0.05)
```

```python
#raw look of our data
plt.scatter(circle_x[:,0],circle_x[:,1],c=circle_y,marker='.')
```

```python
#make non linear model
non_linear=svm.SVC(kernel='rbf',C=1.0)
```

```python
non_linear.fit(circle_x,circle_y)
```

```python
# Plot the decision boundary for a non-linear SVM problem
def plot_decision_boundary(model, ax=None):
    if ax is None:
        ax = plt.gca()
        
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)

	# shape data
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    
	# get the decision boundary based on the model
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary
    ax.contour(X, Y, P,
               levels=[0], alpha=0.5,
               linestyles=['-'])
```

```python
#plit data and decision boundary
plt.scatter(circle_x[:, 0], circle_x[:, 1], c=circle_y, s=50)
plot_decision_boundary(non_linear)
plt.scatter(non_linear.support_vectors_[:, 0], non_linear.support_vectors_[:, 1], s=50, lw=1, facecolors='none')
plt.show()
```
