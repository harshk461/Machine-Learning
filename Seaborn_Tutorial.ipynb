# Seaborn: Heatmap

```python
import seaborn as sns
from matplotlib import pyplot

import numpy as np
```

```python
cars=sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
origin_cyl=(cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 .fillna(0)
)
```

```python
sns.heatmap(origin_cyl)
```

```python
test_array=np.array([[1,2],[3,4]])
```

```python
test_array
```

```python
sns.heatmap(test_array)
```

```python
origin_cyl.T
```

```python
sns.heatmap(origin_cyl.T)
```

# Correlation

```python
cars.corr()
```

```python
sns.heatmap(cars.corr())
```

# Color Palette

```python
sns.heatmap(origin_cyl,cmap="Blues")
```

```python
# Diverging palette
```

```python
sns.heatmap(cars.corr(),cmap='RdBu',center=0,vmin=-1,vmax=1)
```

```python
# Annotations
```

```python
sns.heatmap(origin_cyl,cmap="Blues",annot=True,fmt='.0f')
```

```python
sns.heatmap(
    origin_cyl,
    cmap="Blues",
    annot=True,
    fmt='.0f',
    annot_kws={
        "fontsize":16,
        "fontweight":"bold",
    }
)
```

# Linewidth

```python
sns.heatmap(cars.corr(),cmap='RdBu',linewidth=1)
```

```python
sns.heatmap(cars.corr(),cmap='RdBu',linewidth=1,linecolor="black")
```

```python
labels=['MPG',"Cylinders","Displacement","Horsepower",
        "Weight","Acceleration","Model Year"]
```

```python
sns.heatmap(
    cars.corr(),
    cmap="RdBu",
    xticklabels=labels,
    yticklabels=labels,
)
```

```python
sns.heatmap(
    cars.corr(),
    cmap="RdBu",
    xticklabels=labels,
    yticklabels=labels,
    square=True
)
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins)
```

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
sns.barplot(x='body_mass_g',y='species',data=penguins)
```

# interval Estimates

# confidence intervals(ci)

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            ci=None
           )
```

```python
import numpy as np
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            estimator=np.std
           )
```

```python
penguins.groupby('species').body_mass_g.describe()
```

# Categorical Variables

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            hue='sex'
           )
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            order=['Adelie','Gentoo','Chinstrap'],
            hue='sex',
            hue_order=['Female','Male']
           )
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            color='c',
            errcolor='darkslateblue',
            errwidth=8
           )
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            palette='hot'
)
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            color='aliceblue',
            edgecolor='mediumblue',
            lw=3,
            ci=None
)
```

# Seaborn: Countplot

```python
import seaborn as sns
```

```python
diamonds=sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds=diamonds[diamonds.clarity.isin(['SI1','VS2'])]
```

```python
diamonds
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(y='color',data=diamonds)
```

```python
diamonds.color.value_counts(sort=False)
```

```python
sns.countplot(x='cut',data=diamonds)
```

```python
diamonds.dtypes
```

```python
diamonds.color.cat.categories
```

```python
diamonds.cut.cat.categories
```

```python
color_order=['J','I','H','G','F','E','D']

sns.countplot(x='color',data=diamonds,order=color_order)
```

# order ascending or descending

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(
    x='color',
    data=diamonds,
    order=diamonds.color.value_counts().index[1::],
)
```
